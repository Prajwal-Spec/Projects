```python
# Import Libraries
import tensorflow as tf
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt
import numpy as np
import os
```

Data Collection and Preprocessing

```python
# Path to the root directory containing subfolders for each property class
ds_path = 'REI-Dataset/'

# Parameters
img_height = 160
img_width = 160
batch_size = 32
```

```python
# Load training dataset with 80/20 train/validation split
train_ds = tf.keras.utils.image_dataset_from_directory(
    ds_path,
    validation_split=0.2,
    subset='training',
    seed=123,
    image_size=(img_height, img_width),
    batch_size=batch_size
)
```

```python
class_names = train_ds.class_names
class_names
```

```python
# Load Validation dataset
val_ds = tf.keras.utils.image_dataset_from_directory(
    ds_path,
    validation_split=0.2,
    subset='validation',
    seed=123,
    image_size=(img_height, img_width),
    batch_size=batch_size
)
```

```python
# Normalize pixel values to [0,1] range
normalize_layer = tf.keras.layers.Rescaling(1./255)
train_ds = train_ds.map(lambda x, y: (normalize_layer(x), y))
val_ds = val_ds.map(lambda x, y: (normalize_layer(x), y))
```

```python
# Dataset caching and prefetching for performance optimization
AUTOTUNE = tf.data.AUTOTUNE
train_ds = train_ds.cache().prefetch(buffer_size=AUTOTUNE)
val_ds = val_ds.cache().prefetch(buffer_size=AUTOTUNE)
```

```python
# data augmentation layer
data_augmentation = tf.keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.1),
    layers.RandomZoom(0.1),
])
```

```python
# Verify the data visually
plt.figure(figsize=(10, 10))
for images, labels in train_ds.take(1):
    for i in range(9):
        ax = plt.subplot(3, 3, i + 1)
        plt.imshow((images[i].numpy() * 255).astype('uint8'))
        plt.title(class_names[labels[i]])
        plt.axis('off')
plt.show()
```

Model Selection and Architecture, Transfer Learning

```python
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.applications.resnet import preprocess_input
```

```python
# Base pre-trained model (excluding it's top classifier layer)
base_model = ResNet50(
    weights = 'imagenet',
    input_shape = (160, 160, 3),
    include_top = False
)
base_model.trainable = False # Freeze pre-trained weights for initial training 
```

```python
base_model.summary()
```

```python
# Custom classification head using keras functional API
inputs = layers.Input(shape = (160, 160, 3))
x = data_augmentation(inputs)
x = preprocess_input(inputs) # Use ResNet specific preprocessing
x = base_model(x, training=False) # Recall ResNet50
x = layers.GlobalAveragePooling2D()(x) # 5x5 => GAP => 2048 new features (reduces the feature map into a flat feature vector)
x = layers.Dense(128, activation='relu')(x) # Hidden Layer
x = layers.Dropout(0.4)(x) # to avoid overfitting
outputs = layers.Dense(len(class_names), activation='softmax')(x) # output layer
model = models.Model(inputs, outputs) # New model (transfer learning)
```

```python
# Compile the model
model.compile(
    optimizer = 'adam',
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']
)
```

```python
# Summary
model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

# Callback to improve training
early_stopping = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=2, min_lr=1e-6)

# Train the model
history = model.fit(
    train_ds,
    validation_data = val_ds,
    epochs = 10,
    callbacks = [early_stopping, reduce_lr]
)
```

```python
print('Number of layers in the base model:', len(base_model.layers))
```

Fine-Tuning

```python
# unfreeze the last layers of the base model for fine-tuning
base_model.trainable = True
```

```python
# Fine-tune from this layer onwards
fine_tune_at = 100

# Freeze all the layers before 'fine_tune_at' layer
for layer in base_model.layers[:fine_tune_at]:
    layer.trainable = False
```

```python
inputs = layers.Input(shape=(img_height, img_width, 3))
x = data_augmentation(inputs)
x = preprocess_input(x)
x = base_model(x, training=True) # Set training as True
x = layers.GlobalAveragePooling2D()(x)
x = layers.Dense(128, activation='relu')(x)
x = layers.Dropout(0.4)(x)
outputs = layers.Dense(len(class_names), activation='softmax')(x)

finetune_model = models.Model(inputs, outputs)
```

```python
finetune_model.summary()
```

```python
# Compile the fine-tuned model
finetune_model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-5),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
# Callbacks
early_stopping = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=2, min_lr=1e-6)
```

```python
# Train the fine-tuned model
history_finetune = finetune_model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=10,
    callbacks=[early_stopping, reduce_lr]
)
```

```python
# Learning Curves
plt.plot(history_finetune.history['accuracy'], label='Training Accuracy')
plt.plot(history_finetune.history['val_accuracy'], label='Validation Accuracy')

plt.title('Training vs Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
plt.plot(history_finetune.history['loss'], label='Training Loss')
plt.plot(history_finetune.history['val_loss'], label='Validation Loss')

plt.title('Training vs Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
# Fine-tuning might need more epochs; 10 epochs can still be low for convergence on a complex dataset.
# Train with some additional epochs

# Continue training starting from epoch 10
history_continued = finetune_model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=15,
    initial_epoch=10,
    callbacks=[early_stopping, reduce_lr]
)
```

```python
# Learning Curves
plt.plot(history_continued.history['accuracy'], label='Training Accuracy')
plt.plot(history_continued.history['val_accuracy'], label='Validation Accuracy')

plt.title('Training vs Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
plt.plot(history_continued.history['loss'], label='Training Loss')
plt.plot(history_continued.history['val_loss'], label='Validation Loss')

plt.title('Training vs Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

Model Evaluation

```python
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix

# Generate predictions on validation set
y_true = []
y_pred = []

for images, labels in val_ds:
    preds = finetune_model.predict(images)
    y_true.extend(labels.numpy())
    y_pred.extend(np.argmax(preds, axis=1))
```

```python
# Classification Report
report = classification_report(y_true, y_pred, target_names=class_names)
print('Classification Report:\n', report)
```

```python
# Confusion Matrix
cm = confusion_matrix(y_true, y_pred)

plt.figure(figsize=(8, 8))
sns.heatmap(cm, annot=True, fmt='d', xticklabels=class_names, yticklabels=class_names, cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```
